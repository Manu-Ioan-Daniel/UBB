```python
from random import randint
```

```python
def zile_nastere_sim(n,nr_sim=1000):

  cf,cp=0,0
  i=0
  while(i<nr_sim):
    zile=[randint(0,365) for _ in range(n)]
    zile.sort()
    cp+=1
    for j in range(0,n-1):
      if zile[j]==zile[j+1]:
        cf+=1
        break
    i+=1

  return cf/cp
zile_nastere_sim(30)
```

```python
from matplotlib.pyplot import plot,grid,title,show

```

```python
title('Plot test')
xs=range(1,100)
ys=[zile_nastere_sim(x) for x in xs]
plot(xs,ys,'r*')
grid()
show()
```

```python
from matplotlib.pyplot import axis,plot,show
from math import dist
from random import random
axis('square')
axis((0,1,0,1))
nr_bune=0
for _ in range(10000):
  F=[random(),random()]
  if dist((0.5,0.5),(F[0],F[1]))<0.5:
    plot(F[0],F[1],'ro')
    nr_bune+=1
  else:
    plot(F[0],F[1],'go')
print(nr_bune*4/10000)
show()
```
